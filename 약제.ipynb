일반명코드 - 해당 의약품의 주성분 및 제형에 따라 부여된 코드, 가변문자형(VARCHAR), 9    
제형구분코드 - 의약품을 사용 목적이나 용도에 맞게 적절한 형태로 만든 것, 가변문자형(VARCHAR), 4   
제형 - 의약품을 사용 목적이나 용도에 맞게 적절한 형태로 만든 것, 고정문자형(CHAR), 2   
일반명 - 의약품 주성분의 이름, 가변문자형(VARCHAR), 159   
분류번호 - 해당 의약품의 주성분에 따라 식품의약품안전처에서 분류한 번호, 가변문자형(VARCHAR), 3   
투여경로 - 해당 의약품의 투여 방법(내복, 주사, 외용 등), 고정문자형(CHAR), 10   
함량 - 해당 의약품의 주성분 함량, 가변문자형(VARCHAR), 16    
단위 - 해당 의약품의 주성분 함량의 단위, 가변문자형(VARCHAR), 67   

```python
import pandas as pd

file1 = "C:\Jimin\pharmaLex_unity\data\건강보험심사평가원_약가마스터_의약품주성분_20241014.csv"

df_b = pd.read_csv(file1,encoding='CP949')
df_b.rename(columns={'일반명코드':'주성분코드'}, inplace=True)

df_b.head()
```

```python
# 100101AGN	
# 431631BIJ
# 일반명코드 제형구분코드 제형 일반명 분류번호 투여경로 

file2 = "C:\Jimin\pharmaLex_unity\out\class3_to_product_map.csv"

df_a = pd.read_csv(file2)

merged_df = pd.merge(df_a, df_b, how='inner', on='주성분코드')
```
